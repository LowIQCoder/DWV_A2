```python
import pandas as pd
import plotly.express as px
import os
```

```python
PATH = os.path.abspath('./ip_addresses.csv')

df = pd.read_csv(
    PATH
)
```

```python
df.head(10)
```

```python
df.describe()
```

```python
print("Time Interval:\t\t\t", df['Timestamp'].max() - df['Timestamp'].min())
print("Number of suspicios ips:\t", int(df['suspicious'].sum()))
```

```python
df['Timestamp'] = df['Timestamp'] - df['Timestamp'].min()
df.head(10)
```

```python
fig = px.scatter_geo(
    df,
    lat='Latitude',
    lon='Longitude',
    color='suspicious',
    projection="natural earth",\
    hover_name='ip address'
)
fig.show()
```
